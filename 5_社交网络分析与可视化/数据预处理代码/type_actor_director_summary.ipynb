```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
f = open("movie.csv")
movie = pd.read_csv(f) 
```

# type


```python
type_set = set()  
for x in movie["type"]:  
    type_set.update(str(x).split(',')) 
```

```python
movie["type_cut"] = 0
for i in range(len(movie)):
    movie["type_cut"][i] = str(movie["type"][i]).replace(","," ") 
```

```python
from sklearn.feature_extraction.text import TfidfTransformer  
from sklearn.feature_extraction.text import CountVectorizer 
vectorizer=CountVectorizer()
transformer=TfidfTransformer()
```

```python
tfidf=transformer.fit_transform(vectorizer.fit_transform(movie["type_cut"]))
word=vectorizer.get_feature_names()
weight=tfidf.toarray()
```

```python
for i in range(len(weight)):#打印每类文本的tf-idf词语权重，第一个for遍历所有文本，第二个for便利某一类文本下的词语权重  
    print(u"-------这里输出第",i,u"类文本的词语tf-idf权重------" ) 
    for j in range(len(word)):  
        print(word[j],weight[i][j]) 
```

```python
for i in range(len(movie)):
    movie["type_cut"][i]=list(weight[i])
```

# actors

```python
actors_set = set()  
for x in movie["actors"]:  
    actors_set.update(str(x).split(','))
```

```python
movie["actor_cut"] = 0
for i in range(len(movie)):
    if pd.notnull(movie["actors"][i]):
        movie["actor_cut"][i] = movie["actors"][i].replace(","," ") 
```

```python
def binary2(actor_list):
    binaryList = []
    actor_name = []
    for actor in actors_set:
        if actor in actor_list:
            binaryList.append(1)
        else:
            binaryList.append(0)
        actor_name.append(actor)
    return binaryList,actor_name
```

```python
for i in range(len(movie)): 
    movie["actor_cut"][i],actor_name = binary2(str(movie['actors'][i]))
```

# director

```python
director_set = set()  
for x in movie["directors"]:  
    director_set.update(str(x).split(',')) #870位导演
```

```python
def binary1(director_list):
    binaryList = []
    for director in director_set:
        if director in director_list:
            binaryList.append(1)
        else:
            binaryList.append(0)
    return binaryList
```

```python
movie['director_vec'] = movie['directors'].map(lambda x: binary1(x))
```

```python
from gensim.models import doc2vec
from collections import namedtuple
from nltk import word_tokenize
from gensim.models.doc2vec import TaggedDocument
import gensim
```

```python
from gensim.models.doc2vec import Doc2Vec
model= Doc2Vec.load("d2v.model")
```

```python
stopword= open('哈工大停用词表.txt','r')
input = stopword.readlines()
count = len(input)
print(count)
stopword = set(stopword)
```

```python
import jieba  
def lcut(Intro_movie):
    segment=[]
    segs = jieba.lcut(Intro_movie) # jiaba.lcut()   
    for seg in segs:
        if len(seg)>1 and seg!='\r\n'and (seg not in stopword):          
            segment.append(seg)
    return segment
```

```python
for i in range(len(movie)):
    movie["summary"][i]=list(model.infer_vector( lcut(str(movie["summary"][i])), alpha=0.1, min_alpha=0.0001, steps=5))
```
